```python
stopwords = set(STOPWORDS)
```

```python
grain_wc = WordCloud(
    background_color='lightyellow',
    max_words=2000,
    stopwords=stopwords)

grain_wc.generate(grain)
```

```python
fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 


plt.imshow(grain_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
